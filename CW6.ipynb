For the first part, we implemented a derivative matrix $D$ and verified (thanks Aaron) that $D^2$ is equivalent to taking a second derivative. Notably, to handle the endpoints, we used a single step derivative approximation like $\frac{f(1)-f(0)}{h}$, but this gives error at the endpoints; each consecutive derivative adds an erroneous data point at each end, so in plotting the derivatives we got rid of those. Here are some sample functions plotted with their derivatives:

First derivative of $x^2$ from 0 to 8:

```python
l = calc.vec_central_diff(lambda x: x**2, 0, 8, 100)
x = l[0]
fx = l[1]
dfx = l[2]

plt.plot(x,fx,x,dfx)
plt.legend(["f(x)","f'(x)"])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Here's $\sin{x}$ from 0 to 6, with its second and third derivatives:

```python
l2 = calc.vec_central_diff(lambda x: np.sin(x), 0, 8, 100, 2)
l4 = calc.vec_central_diff(lambda x: np.sin(x), 0, 8, 100, 3)

x = l2[0]
fx = l2[1]
d2fx = l2[2]
d4fx = l4[2]

#for each derivative, you have to cut off one endpoint to get a clean plot
plt.plot(x, fx, x[2:99], d2fx[2:99], x[3:98], d4fx[3:98])
plt.legend(["f(x)","f''(x)","f''''(x)"])
plt.show()
```

For the second part, the integration matrix of a function worked similarly. The endpoints sort of handled themselves; the first endpoint is just zero, leaving the integration needing a constant of integration. The final endpoint (because I did the left trapezoid) has no problems. So for example, here's $\int_0^5 \sin{x} dx = -\cos{x}  \vert_0^5$:

```python
l = calc.vec_trapz(lambda x: np.sin(x),0,5,100)
x = l[0]
fx = l[1]
intfx = l[2]

#need to add constant of integration to your data set
plt.plot(x,fx,x,intfx)
plt.legend(['function','int'])
plt.show()
```

As you can see, the integration is very accurate up to a constant of integration.
